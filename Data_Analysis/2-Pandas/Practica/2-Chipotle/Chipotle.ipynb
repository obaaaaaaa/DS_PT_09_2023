### Paso 2. Importar el conjunto de datos de esta [dirección](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
```

### Paso 3. Asígnalo a una variable llamada chipo.

```python
chipo = pd.read_csv(url, sep = "\t")
chipo.head()
```

# Nombre del producto de valor máximo

```python
chipo["item_price_modified"] = chipo["item_price"].apply(lambda x: float(x[1:]))
chipo["item_price_modified"]
chipo["item_price_modified"].max()
chipo["item_price_individ"] = chipo["item_price_modified"] / chipo["quantity"]
chipo["item_price_individ"].max()
chipo["item_price_individ"]
chipo[chipo["item_price_individ"]==chipo["item_price_individ"].max()].drop_duplicates("item_name")
```

### Paso 4. ¿Cuántos productos cuestan más de 10,00 dólares?

```python
list(chipo[chipo['item_price_individ']>10].sort_values(by="item_price_individ", ascending=False).drop_duplicates("item_name")["item_name"])
```

### Paso 5. ¿Cuál es el precio de cada artículo?
